```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img=cv2.imread('rice.png')

plt.imshow(img)
plt.show()
threshImg=cv2.threshold(img, 100, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(threshImg)
plt.show()

kernel=np.ones((5,5),np.uint8)
ErosionImg=cv2.erode(threshImg,kernel,iterations=1)
DilationImg=cv2.dilate(threshImg,kernel,iterations=1)
DilationErosionImg=cv2.erode(ErosionImg,kernel,iterations=1)

plt.imshow(ErosionImg)
plt.show()
plt.imshow(DilationImg)
plt.show()
plt.imshow(DilationErosionImg)
plt.show()
```
